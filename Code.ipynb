## Self-Attention Mechanism â€“ Example  

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the input sequence  #X - this is the sentence or the sequence that we have  "E.g. I love Sri-lanka " "
input_sequence = np.array([
    [1, 1, 0, 0],  # Token 1
    [0, 0, 0, 0],  # Token 2
    [0, 1, 0, 0]   # Token 3
])

# Initialize random weights for Query, Key, and Value matrices
np.random.seed(42)  # For reproducibility
d_k = input_sequence.shape[1]  # Dimension of embedding
W_Q = np.random.rand(d_k, d_k)
W_K = np.random.rand(d_k, d_k)
W_V = np.random.rand(d_k, d_k)

# try out with weightage of 1 for all
""""ones_matrix = np.ones((4, 4))

W_Q = ones_matrix
W_V = ones_matrix
W_K_q = ones_matrix
"""

# Compute Queries, Keys, and Values using the weight matrix 
Q = np.dot(input_sequence, W_Q)  # (3, 4)
K = np.dot(input_sequence, W_K)  # (3, 4)
V = np.dot(input_sequence, W_V)  # (3, 4)

# Compute the scaled dot-product attention
scores = np.dot(Q, K.T) / np.sqrt(d_k)  # Scale by sqrt(d_k)
attention_weights = np.exp(scores) / np.sum(np.exp(scores), axis=1, keepdims=True)
output = np.dot(attention_weights, V)


# Visualizing Attention Weights
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(attention_weights, cmap="Blues")

# Add labels and a colorbar
ax.set_xticks(range(len(input_sequence)))
ax.set_yticks(range(len(input_sequence)))
ax.set_xticklabels([f"Token {i+1}" for i in range(len(input_sequence))])
ax.set_yticklabels([f"Token {i+1}" for i in range(len(input_sequence))])
ax.set_xlabel("Keys (K)", fontsize=12)
ax.set_ylabel("Queries (Q)", fontsize=12)
ax.set_title("Attention Weights Heatmap", fontsize=14)
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Weight Value")

plt.show()

# Simplified Explanation of the Output
import pandas as pd

# Convert matrices to DataFrames for better readability
attention_df = pd.DataFrame(
    attention_weights,
    columns=[f"Key {i+1}" for i in range(len(input_sequence))],
    index=[f"Query {i+1}" for i in range(len(input_sequence))]
)
output_df = pd.DataFrame(
    output,
    columns=[f"Dim {i+1}" for i in range(output.shape[1])],
    index=[f"Token {i+1}" for i in range(len(input_sequence))]
)

attention_weights,
```
